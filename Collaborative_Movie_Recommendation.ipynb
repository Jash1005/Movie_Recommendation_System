# POPULARITY BASED RECOMENDATION

THE MOST POPULAR 20 MOVIES

```python
# Ensure the movies dataframe has the 'movie_id' column by renaming 'id'
movies.rename(columns={'id': 'movie_id'}, inplace=True)

# Now perform the merge operation
merged = user_ratings.merge(movies, on='movie_id', how='left')
```

```python
merged.head()
```

```python
movie_ratings = merged[['User-ID','Book-Rating','Location','Age','movie_id','title']]
```

```python
movie_ratings.head(1)
```

```python
movie_ratings.groupby('title').count()
```

```python
# Count number of ratings per movie title
movie_rating_counts = movie_ratings.groupby('title').size().reset_index(name='total_num_ratings')

# Show result
print(movie_rating_counts.head())
```

```python
# Ensure 'num-rating' is numeric
movie_ratings['num-rating'] = pd.to_numeric(movie_ratings['num-rating'], errors='coerce')

# Compute average rating per title
average_rating = movie_ratings.groupby('title', as_index=False)['num-rating'].mean()

# Rename column for clarity
average_rating.rename(columns={'num-rating': 'avg-rating'}, inplace=True)

# Show result
print(average_rating.head())
```

```python
movie_rating_list = average_rating.merge(movie_rating_counts, on='title', how='inner')
```

```python
top_20 = movie_rating_list[movie_rating_list['total_num_ratings'] > 250].sort_values('avg-rating', ascending=False).head(20)
```

# COLLABORATIVE

```python
merged.head()
```

only consider those whose total rating are greater than 200 and user has rated more than 50 books

```python
merged.groupby('User-ID').count()['Book-Rating']
```

```python
x = merged.groupby('User-ID').count()['Book-Rating'] > 200
movie_enthusiast_user = x[x].index
#899 user did more than 200 rating
```

```python
filtered_rating = merged[merged['User-ID'].isin(movie_enthusiast_user)]
```

```python
print(filtered_rating.columns)
```

```python
filtered_rating.groupby('title').count()['Book-Rating']
```

```python
x = filtered_rating.groupby('title').count()['Book-Rating'] > 50
famous_movies = x.index
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_movies)]
```

```python
final_ratings.drop_duplicates
```

```python
vectors = final_ratings.pivot_table(index='title',columns='User-ID',values='Book-Rating')

vectors.fillna(0, inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_score = cosine_similarity(vectors)
```

```python
similarity_score
```

```python
def recommend(movie_name) :
   index = np.where(vectors.index==movie_name)[0][0]
   suggestions = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]

   for i in suggestions:
     print(vectors.index[i[0]])
```

```python
recommend('#Horror')
```

```python
recommend('Bride Wars')
```
